```python
import matplotlib.pyplot as plt
labels=['A','B','C','D']
sizes=[15,30,45,10]
plt.pie(sizes,labels=labels)
plt.axis('equal')
plt.show()  
```

```python
import matplotlib.pyplot as plt
products=['Laptop','Desktop','Tablet','Smartphone']
sales=[250000,150,100,30000]
plt.bar(products,sales)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Sales of Products')
plt.show()
```

```python
import matplotlib.pyplot as plt

products = ['Laptop', 'Mobile', 'Tablet', 'TV', 'Headphones']
sales = [50000, 70000, 30000, 40000, 20000]

plt.hist(sales, bins=5, color='orange', edgecolor='black')
plt.xlabel('Sales Range')
plt.ylabel('Number of Products')
plt.title('Histogram of Total Sales for Products')
plt.show()
```

```python
import matplotlib.pyplot as plt
subjects=['Maths','Science','English','History']
marks=[85,90,78,88]
plt.plot(subjects,marks,marker='o')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks in Different Subjects')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
subjects=['Maths','Science','English','History']
marks=[85,90,78,88]
plt.scatter(subjects,marks)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks in Different Subjects')
plt.show()
```
