```python
import pandas as pd
data={
    "Name":["Atharv","vishal","om"],
    "age":[21,22,20]
}
df=pd.DataFrame(data)
print(df)

print("\nhead",df.head(3))

print("\ninfo:",df.info())

print("\ndescribe:",df.describe())

print("\nshape:",df.shape )


print("\nmissing values:",df.isnull().sum())

print("\ncolumns:",df.columns)
```
